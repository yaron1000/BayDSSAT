```python
import hvac
import base64
import json
from google.oauth2 import service_account
from google.cloud import bigquery
import pandas as pd
import os
from serene import adhoc_request, PingCredentials , PagedHandler, async_request, async_retry
import boto3
import io
import sys
import pandas_gbq
import datetime
pd.set_option('display.max_columns', None)

sys.path.append("/mnt/code/CSW/reference")
import functions as fn

def get_secret_from_vault(approle_id, approle_secret, vault_path):
    client = hvac.Client(url='https://vault.agro.services')
    #client.is_authenticated()
    client.auth.approle.login(approle_id, approle_secret)
    return client.read(vault_path)

approle_id = os.environ['APPROLE_ID']
approle_secret=os.environ['APPROLE_SECRET']

secrets = get_secret_from_vault(approle_id = approle_id,
                                approle_secret = approle_secret,
                                vault_path='secret/csw/service-identities/marketdevtrialinglatam') 

if 'data' in secrets and type(secrets['data']['data']) == str:
    service_account_creds = json.loads(base64.b64decode(secrets['data']['data']))
else:
    # in case credentials are saved directly as json object in vault (not encoded) you can get it directly
    service_account_creds = secrets

bq_project = 'bcs-market-dev-lake-np'
bq_credentials = service_account.Credentials.from_service_account_info(service_account_creds)

bq_client = bigquery.Client(project=bq_project, credentials = bq_credentials)

client = boto3.client(
    's3',
    aws_access_key_id= os.environ['aws_key'],
    aws_secret_access_key=os.environ['aws_secret'],
)

query = """
SELECT * FROM `bcs-market-dev-lake-np.latam_datasets.cp_scout_comparison`"""

df = pd.read_gbq(query, project_id=bq_project, credentials=bq_credentials, use_bqstorage_api=True)

df_diff=df[df.diference!=0]

df_diff['TPT-ID']=df_diff['TPT_ID_CSW'].combine_first(df_diff['tpt_id_q'])

df_diff.head(2)

df_diff.loc[df_diff['TPT_ID_CSW'].isnull(),'Error'] = "TPT missing in CSW"

df_diff['protocol_no_CSW'].isnull()

df_diff.loc[df_diff['n_datapointsC']>df_diff['n_datapointsQ'],'Error'] = "More data in CSW"

df_diff.loc[df_diff['n_datapointsC']<df_diff['n_datapointsQ'],'Error'] = "More data in QSR"

df_diff.loc[df_diff['protocol_no_CSW'].isnull(),'Error'] = "Protocol missing in CSW"

df_diff[['TPT-ID','Error']].groupby('Error').count()
```
