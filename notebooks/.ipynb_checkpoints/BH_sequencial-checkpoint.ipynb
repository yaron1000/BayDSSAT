## BH por Thornthwaite e Mather 1955

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Abrindo um arquivo do excel
# ECA ser√° nosso ETP

arquivo=pd.ExcelFile('trabalho_ufu.xlsx')
dados=pd.read_excel(arquivo,'clima_abril')
dados.head()
```

```python
dados['Etc'].plot.line(figsize=(12,3))
etc1=dados['Etc']
plt.savefig('prec_clima_abril.tif',dpi=1000, transparent=True)
```

```python
# Balan√ßo H√≠drico
CAD=15
P=dados['P']
ETC=dados['Etc']

BH=pd.DataFrame({'ANO':dados['ANO'],
                'DJ':dados['NDA'],
                'P':P,
                'ETC':ETC})
BH['P-ETC']=P-ETC

#criar coluna NAC e ARM
BH['NAC']=np.nan
BH['ARM']=np.nan

for dia in range(0,len(BH)):
    if dia==0:
        NACant=0
        ARMant=0
        ARMant=CAD/2
    else:
        n=NACant=BH.loc[dia-1,'NAC']
        ARMant=BH.loc[dia-1,'ARM']
            
    if BH.loc[dia,'P-ETC']<0:
        BH.loc[dia,'NAC']=NACant+BH.loc[dia,'P-ETC']
        BH.loc[dia,'ARM']=CAD*np.exp(BH.loc[dia,'NAC']/CAD)
        if BH.loc[dia,'ARM']>CAD:
            BH.loc[dia,'ARM']=CAD
    else:
        BH.loc[dia,'ARM']=BH.loc[dia,'P-ETC']+ARMant
        if BH.loc[dia,'ARM']>CAD:
            BH.loc[dia,'ARM']=CAD
        BH.loc[dia,'NAC']=CAD*np.log(BH.loc[dia,'ARM']/CAD)

    BH.loc[dia,'ALT']=BH.loc[dia,'ARM']-ARMant
    
    if BH.loc[dia,'ALT']<0:
        BH.loc[dia,'ETR']=BH.loc[dia,'P']+np.abs(BH.loc[dia,'ALT'])
    else:
        BH.loc[dia,'ETR']=BH.loc[dia,'ETC']
    
    BH.loc[dia,'DEF']=BH.loc[dia,'ETC']-BH.loc[dia,'ETR']
    
    if BH.loc[dia,'ARM']<CAD:
        BH.loc[dia,'EXC']=0
    else:
        BH.loc[dia,'EXC']=BH.loc[dia,'P-ETC']-BH.loc[dia,'ALT']
```

```python
BH.head()
```

```python
BHgraf_estrato=pd.DataFrame({'EXC':BH['EXC'],'DEF':-BH['DEF']})
BHgraf_estrato.plot.area(alpha=0.4,cmap='rainbow')
plt.savefig('bh_clima_abril.tif',dpi=1000, transparent=True)
```

```python
BHgraf_estrato=pd.DataFrame({'Precipita√ß√£o':dados['P'],'Irriga√ß√£o':-dados['Irr']})
BHgraf_estrato.plot.area(alpha=0.4,cmap='rainbow')
plt.savefig('pi_clima_abril.tif',dpi=1000, transparent=True)
```
